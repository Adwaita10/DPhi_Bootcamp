# Top 5 East Asian countries based on Generosity


```python
EastAsian = df1[df1['Regional indicator'] == 'East Asia']
```

```python
Generosity=EastAsian.sort_values("Explained by: Generosity",ascending=False).head(5)
Generosity[['Country name','Explained by: Generosity']]
```

# Line Plot of Ladder Score VS top 5 east asian countries based on Generosity

```python
Generosity[['Country name','Ladder score']]
```

```python
Generosity.plot( x='Country name', y='Ladder score',marker='o')
plt.xticks(rotation = 45)     
```
